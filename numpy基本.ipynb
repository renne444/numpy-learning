### 合并分割

```python
x = np.array([1,2,3])
y = np.array([3,2,1])
z = np.array([666, 666, 666])
```

```python
np.concatenate([x, y, x])
```

```python
A = np.array([x, y])
A
```

```python
np.concatenate([A, A])
```

```python
np.concatenate([A, A], axis=1)
```

```python
A2 = np.concatenate([A, z.reshape(1, -1)])
A2
```

```python
np.vstack([A, z])
```

```python
B = np.full((2,2), 100)
B
```

```python
np.hstack([A, B])
```
