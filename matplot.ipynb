```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, siny, label="sinx")
plt.plot(x, cosy, color="red", linestyle="--", label="cosx")
plt.xlim(5, 10)
plt.xlabel("x-label")
plt.ylabel("sin-x")
plt.legend()
plt.show()
```

```python
plt.scatter(x, siny, label="sinx")
plt.plot(x, cosy, color="red", linestyle="--", label="cosx")
plt.xlim(5, 10)
plt.xlabel("x-label")
plt.ylabel("sin-x")
plt.legend()
plt.show()
```
